```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("BM APP").getOrCreate()
print(spark.version)
rdd=spark.sparkContext.parallelize([1,2,3,4,5])
print(rdd.collect())
```

```python
squared_rdd=rdd.map(lambda x:x*x)
print(squared_rdd.collect())
```

```python
data = [("Bharath1", 19,"ece","a",1),("Bharath1", 20,"ece","b",1),("Bharath", 21,"ece","c",2),("Bharath", 22,"ece","d",7),("Bharath", 23,"ece","e",5),("Bharath", 24,"ece","f",1),("Bharath", 25,"ece","g",3),("Bharath", 26,"ece","h",14),]
columns = ["Name", "AGE","DEPT","SEC","RANK"]
df = spark.createDataFrame(data, columns)
df.show()
```

```python
from pyspark.sql.functions import col
df.withColumn("passed YEAR",2046-col("AGE")).show()
```

```python
df.groupBy("RANK").count().show()
df.filter(col("RANK")<5).show()
```
