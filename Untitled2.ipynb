```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("MOVIE RATINGS").getOrCreate()
print(spark.version)
```

```python
tmovies_df=spark.read.csv("/content/tamimovie.csv", header=True,inferSchema=True)
ratings_df=spark.read.csv("/content/rating.csv",header=True,inferSchema=True)
tmovies_df.show(10)
ratings_df.show(10)
```

```python
from pyspark.sql.functions import avg
avg_ratings_df=ratings_df.groupBy("movieId").agg(avg("rating").alias("avg_rating"))
avg_ratings_df.show()
```

```python
from pyspark.sql.functions import count
rating_counts_df=ratings_df.groupBy("movieId").agg(count("rating").alias("num_ratings"))
rating_counts_df.show()
```

```python
ratings_summary_df=avg_ratings_df.join(rating_counts_df,"movieId")
ratings_with_titles=ratings_summary_df.join(tmovies_df,"movieId")
ratings_with_titles.select("title","avg_rating","num_ratings").orderBy("num_ratings",ascending=False).show()
```
