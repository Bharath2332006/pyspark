```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("newone").getOrCreate()
print(spark.version)
```

```python
data=[("Bharath",25,"ece"),("barath",19,"ece"),("bharat",21,"ece")]
columns =["Name","age"]
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df.select("Name").show()
df.filter(df["age"]>19).show()
```
