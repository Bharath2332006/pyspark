```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("csvcreate").getOrCreate()
print(spark.version)
data=[("BM",19,"java,html,css"),("DK",20,"java"),("JP",19,"java"),("RM",20,"java"),("TG",20,"java")]
columns=["name","age","skill"]
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df.write.csv("csvcreate.csv",header=True)
```

```python
df.groupBy("age").count().show()
```

```python
df.groupBy("skill").count().show()
```

```python
df.createOrReplaceTempView("age")
spark.sql("SELECT * FROM age WHERE AGE >=19").show()
```
