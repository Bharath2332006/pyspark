```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('BHARATH APP').getOrCreate()
df = spark.read.csv("/content/spotify-data.csv", header=True, inferSchema=True)
df.filter(df["release_date"]>2006).show(5)
```

```python
df.filter((df["loudness"]>0.1)& (df["energy"]>0.5)&(df["artist"]==)).select("name","loudness","energy","artists").show(5)
```

```python
df.select("name", "artists", "popularity").orderBy("popularity", ascending=True).show(10)
```

```python
df.select("name", "loudness").orderBy("loudness", ascending=False).show(10)
```

```python
from pyspark.sql.functions import col, isnan, when, count

df.select([count(when(col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```
